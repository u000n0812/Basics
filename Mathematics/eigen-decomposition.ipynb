# Eigenvalue / Eigenvector

**Ax** = **cx**  

**x** : eigenvector  
**c** : eigenvalue  
**A** : transformation matrix

Eigen decomposition :  

**X** = **VDV'**  

**D** : Diagonal matrix with eigenvalue  -> "Diagonalizble"  
**V** : matrix with eigenvector

*Property of eigen-decomposition*

If **A** is symmetric, then **A** is diagonalizable.  

Also, **A** can be decomposed to **QDQ'**, where each column of **Q** is orthogonable with each other.  

Then, **A** can be written as the product of (*eigenvalue*,  *column of Q*)  

This means matrix **A** is composed of some layers lambda_1, lambda_2, ... , lambda_n. If we use only some part of them, it is called **'dimension reduction'**

```python
Image("C:/Users/jy990/Study/etc/eigen-decomposition을 하는 이유.PNG")
```

As a result, If **A** is symmetric matrix, it can be eigen-decomposition.  

In this view, **Ax** can be considered to the combination of vectors which were made through eigen-decomposition.  

vector **x** is projected to the orthonormal eigenvectors q_1, q_2, q_3, ... , q_n. And, the combination of the product of (projection, d_1, d_2, d_3, ... , d_n) is exactly the same as *linear transformation* **Ax**

Plus, there is a more awesome method which does not need any restrictions such as "**A** must be symmetric"...  

This method is called *SVD-decomposition*
