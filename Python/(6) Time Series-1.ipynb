```python
import pandas as pd

dat_df = pd.read_csv('C:/Users/jy990/OneDrive/바탕 화면/뇌/2022/2022/Python/머신러닝데이터 전처리입문/데이터셋/chap3_4_6_7_A1-5/chap6/data/energydata.csv', sep=',')
dat_df.head()
```

```python
print(dat_df.shape)
```

```python
print(dat_df.dtypes)
```

```python
dat_df['date'] = pd.to_datetime(dat_df['date'], format='%Y-%m-%d %H:%M:%S') #date 포맷 변경
print(dat_df['date'].dtypes)
print(type(dat_df['date'][0]))
```

```python
dat_df['dif_min'] = dat_df['date'].diff().dt.total_seconds()/60 # diff로 한 행 전 값과의 차이 계산한 뒤 초 단위의 값을 구해 60으로 나눈다
dat_df['dif_min'] = dat_df['dif_min'].fillna(0) #결손값은 0으로 보완한다.
dat_df['dif_min'].head()
```

```python
dat_df['cum_min'] = dat_df['dif_min'].cumsum()
dat_df[['date', 'cum_min']].head()
```

```python
dat_df['cum_hour'] = (dat_df['cum_min']/60).round(2).astype(int) # 시간 단위의 경과시간 게산
dat_df[['date', 'cum_min', 'cum_hour']].head(10)
```

```python
dat_df.groupby('cum_hour').mean() # 경과시간별로 평균치 구함
```

```python
dat_df.groupby('cum_hour').std()
```

```python
print(dat_df.isnull().sum(axis=1).sort_values(ascending=False))
print(dat_df.isnull().sum(axis=0))
```
