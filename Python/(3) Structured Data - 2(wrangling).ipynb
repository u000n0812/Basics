```python
import pandas as pd

bank_df = pd.read_csv('C:/Users/jy990/OneDrive/바탕 화면/뇌/2022/2022/Python/머신러닝데이터 전처리입문/데이터셋/chap3_4_6_7_A1-5/chap3-4/data/bank.csv', sep=',')
bank_df.head()
```

```python
print(bank_df.shape)
print(bank_df.dtypes)
```

```python
bank_df = bank_df.dropna(subset=['job', 'education']) # job과 education은 결손값이 별로 없으므로 NA값을 그냥 삭제해준다.
print(bank_df.shape)
```

```python
bank_df = bank_df.dropna(thresh=2400, axis=1) # 결손값이 2400보다 큰 열을 삭제한다.
print(bank_df.shape)
```

```python
bank_df = bank_df.fillna({'contact' : 'unknown'}) # contact의 NA값을 unknown으로 교체한다.
bank_df.head()
```

```python
bank_df = bank_df[bank_df['age'] >= 18]
bank_df = bank_df[bank_df['age'] < 100] # 나이가 18세 이상 100세 미만인 사람들만 남긴다.

print(bank_df.shape)
```

```python
bank_df = bank_df.replace(['yes', 'no'], ['1', '0']) # 문자형 데이터를 수치형을 바꾸기 위해 yes는 1로, no는 0으로 바꾼다.

bank_df.head()
```

```python
bank_df_job = pd.get_dummies(bank_df['job']) # 다수 값을 취하는 job 변수에 대해서는 더미변수화한다. 

bank_df_job.head()
```

```python
bank_df_marital = pd.get_dummies(bank_df['marital'])
bank_df_education = pd.get_dummies(bank_df['education'])
bank_df_contact = pd.get_dummies(bank_df['contact'])
bank_df_month = pd.get_dummies(bank_df['month'])
```

```python
tmp1 = bank_df[['age', 'default', 'balance', 'housing', 'loan', 'day', 'duration', 'campaign', 'pdays', 'previous', 'y']]

tmp1.head()
```

```python
tmp2 = pd.concat([tmp1, bank_df_marital], axis=1) # concat 함수로 2개의 데이터를 결합한다. axis=1을 사용해 수평으로 결합한다.
tmp3 = pd.concat([tmp2, bank_df_education], axis=1)
tmp4 = pd.concat([tmp3, bank_df_contact], axis=1)
bank_df_new = pd.concat([tmp4, bank_df_month], axis=1)

bank_df_new.head()
```

```python
bank_df_new.to_csv('bank-prep.csv', index=False) # 데이터를 CSV 파일로 출력한다.
```
