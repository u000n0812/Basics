```python
import cv2 # 이미지 데이터를 처리하기 위한 OpenCV 호출
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

img = np.fromfile('C:/Users/jy990/OneDrive/바탕 화면/뇌/2022/2022/Python/머신러닝데이터 전처리입문/데이터셋/chap5_source/Chap5_Source/data/ants/swiss-army-ant.jpg')
# 바로 이미지가 읽히지 않아서 numpy의 fromfile로 읽었다.
img = cv2.imdecode(img, cv2.IMREAD_COLOR)
# numpy로 읽었기 때문에 OpenCV에서 읽을 수 있도록 변환시켜준다.

print(img.shape)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # cvtColor를 이용해 픽셀값의 정렬를 RGB(적,녹,청)로 바꾸고 imshow로 시각화한다.
plt.show()
```

### 이미지에는 컬러를 표하는 0~255 사이의 RGB 값 뿐 아니라 밝기의 정도(휘도)만으로 표현하는 그레이스케일 이미지도 있다. 모델 작성을 위한 특징량으로 휘도 픽셀값을 사용하고 싶으면 이미지를 그레이스케일 이미지로 변환해야 한다.


```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # RGB 픽셀 값을 휘도 픽셀 값으로 변환한다.

print(gray_img.shape)

plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
gr_df = pd.DataFrame(gray_img)

print(gr_df.shape)
gr_df.head()
```

### 그레이스케일 이미지보다 정보를 더 줄이고 특징량을 돋보이게 하는 것이 이진화(binarization) 이미지이다. 이진화 이미지는 픽셀값이 경계값보다 크면 255, 작으면 0으로 주어 흑백 이미지로 변환한다.

```python
ret, bin_img = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY)

plt.imshow(bin_img, cmap='gray')
plt.show()
```

```python
print(bin_img)
```

```python
bin_df = pd.DataFrame(bin_img)

print(bin_df.shape)
bin_df.head()
```
