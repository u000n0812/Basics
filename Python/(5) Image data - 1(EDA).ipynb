```python
import cv2 #     OpenCV 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

img = np.fromfile('C:/Users/jy990/OneDrive/ //2022/2022/Python/ //chap5_source/Chap5_Source/data/ants/swiss-army-ant.jpg')
#     numpy fromfile .
img = cv2.imdecode(img, cv2.IMREAD_COLOR)
# numpy   OpenCV    .

print(img.shape)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # cvtColor    RGB(,,)  imshow .
plt.show()
```
